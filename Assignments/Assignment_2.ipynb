13. Find outliers if there is any how would you handle them

```python
## Your code here down

import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = df.select_dtypes(include='number').columns
num_plots = len(numerical_columns)
num_rows = (num_plots + 1) // 2  # Calculate the number of rows needed
fig, axes = plt.subplots(num_rows, 2, figsize=(12, 6)) 
for i, column in enumerate(numerical_columns):
    row = i // 2
    col = i % 2
    sns.boxplot(x=df[column], ax=axes[row, col])
    axes[row, col].set_title(column)

plt.tight_layout()
plt.show()


def count_outliers_iqr(dataframe: pd.DataFrame, threshold: float = 1.5) -> dict:
    """
    Count outliers in each numerical column of a DataFrame using the Interquartile Range (IQR) method.

    Args:
        dataframe (pd.DataFrame): The input DataFrame.
        threshold (float, optional): The threshold for detecting outliers. Default is 1.5.

    Returns:
        dict: A dictionary where keys are column names and values are the counts of outliers for each column.

    Raises:
        ValueError: If the DataFrame contains no numerical columns.

    """
    # Check if the DataFrame is empty or contains no numerical columns
    if dataframe.empty or not any(dataframe.select_dtypes(include='number')):
        raise ValueError("DataFrame is empty or contains no numerical columns.")

    # Initialize a dictionary to store outlier counts for each column
    outlier_counts = {}

    # Loop through each numerical column in the DataFrame
    for column in dataframe.select_dtypes(include='number').columns:
        # Calculate the first quartile (Q1) and third quartile (Q3)
        Q1 = dataframe[column].quantile(0.25)
        Q3 = dataframe[column].quantile(0.75)

        # Calculate the Interquartile Range (IQR)
        IQR = Q3 - Q1

        # Define the lower and upper bounds for outliers
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR

        # Count the number of outliers in the column
        outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)]
        outlier_count = len(outliers)

        # Store the outlier count in the dictionary
        outlier_counts[column] = outlier_count

    return outlier_counts

print("Display # of Outliers per Column:\n",count_outliers_iqr(df))
print ("No outliers detected. \nIf outliers existed they could be resolved by removing, transforming, setting to predefined threshold, or imputation.")


## Your code here up
```

14. Find descriptive statistics for relavant columns and plot them by using bars.

```python
## Your code here down

from typing import List

#Descriptive Statistics for Numerical Data
stats_numerical = df.describe()
print(stats_numerical)

#Descriptive Statistics for Categorical Data
stats_categorical = df[['farm_name', 'soil_type']].describe()
print(stats_categorical)

# graph the data to visualize patterns
def draw_histograms(dataframe: pd.DataFrame, features: List[str], rows: int, cols: int):
    """
    Draw histograms for specified features of a DataFrame.

    Args:
        dataframe (pd.DataFrame): The DataFrame containing the data to visualize.
        features (List[str]): A list of column names to draw histograms for.
        rows (int): Number of rows in the subplot grid.
        cols (int): Number of columns in the subplot grid.

    Returns:
        None
    """
    try:
        # Check if the number of subplots is greater than the number of features
        if len(features) > rows * cols:
            raise ValueError("Number of subplots should be greater than or equal to the number of features.")
        
        fig = plt.figure(figsize=(20, 20))
        
        for i, feature in enumerate(features):
            ax = fig.add_subplot(rows, cols, i + 1)
            dataframe[feature].hist(bins=10, ax=ax, facecolor='midnightblue')
            ax.set_title(f'{feature} Distribution', color='DarkRed')
        
        fig.tight_layout()
        plt.show()
    except Exception as e:
        print(f"An error occurred: {e}")


def visualize_numeric_columns(dataframe: pd.DataFrame, relevant_columns: List[str]):
    """
    Visualize histograms, box plots, and summary statistics for numeric columns in a DataFrame.

    Args:
        dataframe (pd.DataFrame): The DataFrame containing the data to visualize.
        relevant_columns (List[str]): A list of column names to visualize.

    Returns:
        None
    """
    # Filter out non-numeric columns
    numeric_columns = [col for col in relevant_columns if pd.api.types.is_numeric_dtype(dataframe[col])]

    # Create subplots for histograms, box plots, and bar charts
    num_rows = len(numeric_columns)
    num_cols = 3  # Each column for histogram, box plot, and summary stats

    fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 5 * num_rows))

    for i, column in enumerate(numeric_columns):
        # Plot histogram
        sns.histplot(data=dataframe, x=column, ax=axes[i, 0], kde=True)
        axes[i, 0].set_title(f'{column} Histogram')
        
        # Plot box plot
        sns.boxplot(data=dataframe, y=column, ax=axes[i, 1])
        axes[i, 1].set_title(f'{column} Box Plot')
        
        # Plot bar chart for summary statistics
        stats = dataframe[column].describe()
        sns.barplot(x=stats.values, y=stats.index, ax=axes[i, 2])
        axes[i, 2].set_title(f'{column} Summary Stats')

        # Annotate the bars with statistics values
        for p, value in zip(axes[i, 2].patches, stats.values):
            axes[i, 2].annotate(f'{value:.2f}', (p.get_x() + p.get_width(), p.get_y() + p.get_height() / 2),
                           ha='left', va='center', fontsize=10, color='black')
        axes[i, 2].set_title(f'{column} Summary Stats')

    # Adjust layout
    plt.tight_layout()
    plt.show()


draw_histograms(df,df.columns,8,4) # this includes output of all columns

# visualize all numeric data
visualize_numeric_columns(df, ['temperature', 'precipitation', 'percentage_tot_prec', 'precipitation_rank'])

## Your code here up
```
